```python
# Robjectの展開
robj = robjects.r['load']('./data/.RData')
```

```python
# データをデータフレーム化
# numpyにしてから転置をする必要がある。
stock_price = pd.DataFrame(np.array(robjects.r['price4']).T, columns=['板硝子', 'ヤマハ', 'ポーラHD', '武田薬品'])
```

```python
stock_price.plot(kind='line', subplots=True, figsize=(8,8), layout=(2, 2))
```

```python
(stock_price / stock_price.iloc[0, :]).plot(kind='line', subplots=True, figsize=(8,8), layout=(2,2))
```

```python
(np.log(stock_price).diff() * 100).plot(kind='line', subplots=True, layout=(2,2), figsize=(15,8), sharey=True)
```

```python
(np.log(stock_price).diff() * 100).mean()
```
